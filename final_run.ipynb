### Generating Data
Here you can add more data to the current simulation counts. This function plays and records a specified number of new games by generating randomized decks of cards and determining which sequence would win in each theoretical match up. Change the number in play_n_games to alter how many games you want to run.
- Ex. results = main.play_n_games(5, 'data') -> adds 5 additional simulations. 

Remember that however many simulations you run will be added to any data that already exists in the data folder. 

```python
results = main.play_n_games(1000, 'data')
```

### Generating Heatmaps
There are two functions which can be used to generate heatmaps that visualize the simulation results. One generates a single heatmap for games scored either by number of cards won or by number of tricks won. The other one generates a figure showing both heatmaps side by side. 
- create_heatmap requires a string input which can be either 'cards' or 'tricks' depending on which variation you would like to see. This function creates a png and html file of the heatmap, both of which automatically save to the figures folder. 
- make_heatmap_package() does not require any inputs. It generates one visual that shows side-by-side heatmaps representing wins by cards and by tricks. Like the create_heatmap function, it creates a png and html file of the heatmap, both of which automatically save to the figures folder. 

```python
main.create_heatmap('cards')
```

```python
main.create_heatmap('tricks')
```

```python
main.make_heatmap_package()
```
