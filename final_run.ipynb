### Generating Data
Here you can add more data to the current simulation counts. This function plays and records a specified number of new games by generating randomized decks of cards and determining which sequence would win in each theoretical match up. Change the number in play_n_games to alter how many games you want to run.
- Ex. results = main.play_n_games(5, 'data') -> adds 5 additional simulations. 

Remember that however many simulations you run will be added to any data that already exists in the data folder. 

```python
results = main.play_n_games(10, 'data')
```

## Generating Heatmaps
There are two functions that can be used to generate heatmaps to visualize the simulation results. One generates one heatmap for either games scored by number of cards won or number of tricks one. The other one generates heatmaps for games scored by both cards and by tricks side by side. 
- create_heatmap requires of input of what variation you are using, so you must enter the string either 'cards' or 'tricks' depending on which heatmap you want to generate. This function also creates a png and html file of the heatmap that will automatically save to the figures folder. 
- make_heatmap_package() does not require any inputs and generates one visual that shows both heatmaps. You simply need to call the function to get it to run properly. Like the create_heatmap function, it also creates a png and html file of the heatmap that will automatically save to the figures folder. 

```python
main.create_heatmap('cards')
```

```python
main.create_heatmap('tricks')
```

```python
main.make_heatmap_package()
```
